```python
def printst():
    s=int(input())
    st=[]
    for i in range(s):
        op=input().split()
        if op[0]=="push":
            st.append(int(op[1]))
        elif op[0]=="pop":
            st.pop()
        elif op[0]=="peak":
            print(st[-1])
        elif op[0]=="maximum":
            print(max(st))
        elif op[0]=="minimum":
            print(min(st))
        elif op[0]=="printstack":
            print(st)
printst()
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class Stack:
    def __init__(self):
        self.top=None
    def push(self,x):
        n=Node(x)
        n.next=self.top
        self.top=n
    def pop(self):
        if self.top:
            self.top=self.top.next
    def peak(self):
        if self.top:
            print(self.top.data)
    def maximum(self):
        if self.top:
            t=self.top
            m=t.data
            while t:
                m=max(m,t.data)
                t=t.next
            print(m)
    def minimum(self):
        if self.top:
            t=self.top
            m=t.data
            while t:
                m=min(m,t.data)
                t=t.next
            print(m)
    def printstack(self):
        res=[]
        t=self.top
        while t:
            res.append(t.data)
            t=t.next
        print(res)
def printst():
    s=Stack()
    n=int(input())
    
    for i in range(n):
        op=input().split()
        
        if op[0]=="push":
            s.push(int(op[1]))
        elif op[0]=="pop":
            s.pop()
        elif op[0]=="peak":
            s.peak()
        elif op[0]=="maximum":
            s.maximum()
        elif op[0]=="minimum":
            s.minimum()
        elif op[0]=="printstack":
            s.printstack()

printst()
        
```
