```python
def sortst(l,r):
    res=[]
    i=j=0
    while i<len(l) and j<len(r):
        if len(l[i])<len(r[j]):
            res.append(l[i])
            i+=1
        else:
            res.append(r[j])
            j+=1
    if i<len(l):
        res.append(l[i])
        i+=1
    if j<len(r):
        res.append(r[j])
        j+=1
    return res
def merge(a):
    if len(a)<=1:
        return a
    m=len(a)//2
    l=a[:m]
    r=a[m:]
    return sortst(l,r)
s=input().split()
print(merge(s))
```
