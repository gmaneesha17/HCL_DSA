```python
def countoccu(s,a):
    n=len(s)
    m=len(a)
    c=0
    for i in range(n-m+1):
        m=True
        for j in range(m):
            if s[i+j]!=a[j]:
                m=False
        if m:
            c=c+1
    return c
s=input("Enter String")
a=input("Enter sub-string")
print(countoccu(s,a))
```

```python
def repldig(s):
    r=""
    for ch in s:
        if ch >='0' and ch<="9":
            r+="#"
        else:
            r+=ch
    return r
s=input()
print(repldig(s))
```

```python
def sqev(s):
    n=len(s)
    r=[]
    for i in range(n):
        if s[i]%2==0:
            r.append(s[i]*s[i])
        else:
            r.append(s[i]*s[i]*s[i])
    return r
n=int(input())
a=list(map(int,input().split()))
print(sqev(a))
```
