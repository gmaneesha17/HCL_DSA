```python
def factorization(n):
    freq={}
    while n%2==0:
        if 2 in freq:
            freq[2]+=1
        else:
            freq[2]=1
        n=n//2
    i=3
    while i*i<=n:
        while n%i==0:
            if i in freq:
                freq[i]+=1
            else:
                freq[i]=1
            n=n//i
        i=i+2
    if n>1:
        if n in freq:
            freq[n]+=1
        else:
            freq[n]=1
    return freq
def factors(freq):
    r=[]
    for i in sorted(freq):
        if freq[i]>1:
            r.append(str(i)+"^"+str(freq[i]))
        else:
            r.append(str(i))
    return "*".join(r)
n=int(input())
freq=factorization(n)
print(factors(freq))
```
