```python
#abstract class- contains normal methods, abstract methods
from abc import ABC, abstractmethod
class ATM(ABC):
    def show(self):
        print("ATM")
    @abstractmethod
    def pay(self):
        pass
class phonepe(ATM):
    def pay(self):
        print("Payed- Phonepe")
a=phonepe()
a.pay()
a.show()
```

```python
#interface-contains only abstract methods
from abc import ABC, abstractmethod
class ATM(ABC):
    @abstractmethod
    def pay(self):
        pass
class phonepe(ATM):
    def pay(self):
        print("Payed- Phonepe")
a=phonepe()
a.pay()
```

```python
#packing-*args- mutliple values in one variable
def add(*nums):
    print(nums)
add(1,2,3)
```

```python
#unpacking- separate variables
a=[1,2,3]
def add(x,y,z):
    print(x,y,z)
add(*a)
```

```python
#packing-**kwargs
def show(**data):
    print(data)
show(name="Mani",age=21)
```

```python
#unpacking-**kwargs:
data={'name': 'Mani', 'age': 21}
def show(name,age):
    print(name,age)
show(**data)
```

```python
#pickling_converting objects in to byte stream  - save object
import pickle
data={'name': 'Mani', 'age': 21}
with open('file.pkl',"wb") as f:
    pickle.dump(data,f)
```

```python
#unpickle - load object
import pickle
with open('file.pkl',"rb") as f:
    data=pickle.load(f)
print(data)
```
