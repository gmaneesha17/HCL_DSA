```python
def event(a):
    freq={}
    for ch in a:
        if ch in freq:
            freq[ch]+=1
        else:
            freq[ch]=1
    me=None
    mf=-1
    for k in freq:
        if freq[k]>mf:
            mf=freq[k]
            me=k
    return freq,me
a=input().split()
freq,me=event(a)
for k in freq:
    print(k, ":", freq[k])
print("Common element:",me)
```
