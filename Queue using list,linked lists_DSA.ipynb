```python
def printq():
    qu=int(input())
    q=[]
    for i in range(qu):
        op=input().split()
        if op[0]=="enqueue":
            q.append(int(op[1]))
        elif op[0]=="dequeue":
            q.pop(0)
        elif op[0]=="maximum":
            print(max(q))
        elif op[0]=="minimum":
            print(min(q))
        elif op[0]=="peak":
            print(q[0])
        elif op[0]=="printqueue":
            print(q)
printq()
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class queue:
    def __init__(self):
        self.front=None
        self.rear=None
    def enqueue(self,x):
        n=Node(x)
        if self.front is None:
            self.front=n
            self.rear=n
        else:
            self.rear.next=n
            self.rear=n
    def dequeue(self):
        if self.front:
            self.front=self.front.next
            if self.front is None:
                self.rear=None
    def peak(self):
        if self.front:
            print(self.front.data)
    def maximum(self):
        if self.front:
            t=self.front
            m=t.data
            while t:
                m=max(m,t.data)
                t=t.next
            print(m)
    def minimum(self):
        if self.front:
            t=self.front
            m=t.data
            while t:
                m=min(m,t.data)
                t=t.next
            print(m)
    def printqueue(self):
        res=[]
        t=self.front
        while t:
            res.append(t.data)
            t=t.next
        print(res)
def printq():
    q=queue()
    n=int(input())

    for i in range(n):
        op=input().split()

        if op[0]=="enqueue":
            q.enqueue(int(op[1]))
        elif op[0]=="dequeue":
            q.dequeue()
        elif op[0]=="peak":
            q.peak()
        elif op[0]=="maximum":
            q.maximum()
        elif op[0]=="minimum":
            q.minimum()
        elif op[0]=="printqueue":
            q.printqueue()

printq()
```
