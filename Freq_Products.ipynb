```python
def frprod(a):
    freq={}
    for ch in a:
        if ch in freq:
            freq[ch]+=1
        else:
            freq[ch]=1
    d=[]
    for k in freq:
        if freq[k]>1:
            d.append(k)
    return d
a=list(map(int,input().split()))
r=frprod(a)
if r:
    print("Duplicate IDS's :", r)
else:
    print("No Duplicates")
```
