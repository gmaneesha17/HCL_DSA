```python
def lower(w):
    r=""
    for ch in w:
        if 'A'<=ch<='Z':
            r+=chr(ord(ch)+32)
        else:
            r+=ch
    return r
def unique(w):
    u=[]
    for ch in w:
        ch=lower(ch)
        found=False
        for i in u:
            if ch==i:
                found=True
        if not found:
            u.append(ch)
    for i in range(len(u)):
        for j in range(i+1,len(u)):
            if u[i]>u[j]:
                u[i],u[j]=u[j],u[i]
    return u
s=input()
w=s.split()
print(unique(w))
```

```python
def merge(d1,d2):
    r={}
    for key in d1:
        r[key]=d1[key]
    for key in d2:
        if key in r:
            r[key]+=d2[key]
        else:
            r[key]=d2[key]
    return r
n=int(input())
d1={}
for i in range(n):
    key=input("Enter key")
    value=int(input("Enter Quantity"))
    d1[key]=value
d2={}
for i in range(n):
    key=input("Enter key")
    value=int(input("Enter Quantity"))
    d2[key]=value
print(merge(d1,d2))
    
```

```python
def count(a):
    r=[]
    v=[]
    for n in a:
        found=False
        for i in v:
            if n==i:
                found=True
        if not found:
            c=0
            for x in a:
                if n==x:
                    c+=1
            r.append((n,c))
            v.append(n)
    return r
a=list(map(int,input().split()))
print(count(a))
```

```python
def word_frequency(p):
    w=p.split()
    r={}
    for ch in w:
        if ch in r:
            r[ch]+=1
        else:
            r[ch]=1
    sf={}
    while r:
        mw=None
        mc=-1
        for key in r:
            if r[key]>mc:
                mc=r[key]
                mw=key
        sf[mw]=mc
        del r[mw]
    return sf
    
s=input()
print(word_frequency(s))
```

```python
def high_sal(d1):
    me=""
    ms=0
    for key in d1:
        if d1[key]>ms:
            ms=d1[key]
            me=key
    return me,ms
n=int(input())
d1={}
for i in range(n):
    key=input("Enter name")
    value=int(input("Enter salary"))
    d1[key]=value
print(high_sal(d1))
```

```python
def group_category(d):
    r={}
    for i,c in d:
        if c in r:
            r[c].append(i)
        else:
            r[c]=[i]
    return r
    
n=int(input())
d=[]
for i in range(n):
    key=input("Enter item")
    value=input("Enter category")
    d.append((key,value))
print(group_category(d))
```

```python
def fib(n,cache={}):
    if n in cache:
        return cache[n]
    if n==0:
        return 0
    if n==1:
        return 1
    cache[n]=fib(n-1,cache)+fib(n-2,cache)
    return cache[n]
n=int(input())
print(fib(n))
```

```python
def memoize(func):
    cache={}
    def wrapper(n):
        if n in cache:
            return cache[n]
        r=func(n)
        cache[n]=r
        return r
    return wrapper
@memoize
def fib(n):
    if n==0:
        return 0
    if n==1:
        return 1
    return fib(n-1)+fib(n-2)
n=int(input())
print(fib(n))
```
