```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
def createlist(a):
    h=None
    t=None
    for i in a:
        new=Node(i)
        if h is None:
            h=new
            t=new
        else:
            t.next=new
            t=new
    return h
def merge(l1,l2):
    d=Node(0)
    t=d
    while l1 and l2:
        if l1.data<l2.data:
            t.next=l1
            l1=l1.next
        else:
            t.next=l2
            l2=l2.next
        t=t.next
    if l1:
        t.next=l1
    if l2:
        t.next=l2
    return d.next
def printlist(h):
    while h:
        print(h.data)
        h=h.next
    print("None")
a=list(map(int,input().split()))
b=list(map(int,input().split()))
l1=createlist(a)
l2=createlist(b)
h=merge(l1,l2)
printlist(h)
```
