```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None
def create(a):
    h=None
    t=None
    for i in a:
        new=Node(i)
        if h is None:
            h=new
            t=new
        else:
            t.next=new
            new.prev=t
            t=new
    return h
def tsum(h,t):
    res=[]
    l=h
    r=h
    while r.next:
        r=r.next
    while l!=r and l.prev!=r:
        s=l.data+r.data
        if s==t:
            res.append((l.data,r.data))
            l=l.next
            r=r.prev
        elif s<t:
            l=l.next
        else:
            r=r.prev
    return res
def printlist(h):
    while h:
        print(h.data,end=" ")
        h=h.next
    print("None")
a=list(map(int,input().split()))
t=int(input())
h=create(a)
printlist(h)
pairs=tsum(h,t)
if pairs:
    for i in pairs:
        print(i)
else:
    print("Not found")
        
```
