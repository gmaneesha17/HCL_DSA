```python
def count(n):
    i=1
    while i<=n:
        yield i
        i+=1
n=int(input())
for i in count(n):
    print(i)
```

```python
def fib(n):
    c=0
    a=0
    b=1
    while c<n:
        yield a
        a,b=b,a+b
        c+=1
n=int(input())
for c in fib(n):
    print(c,end=" ")
```

```python
from itertools import islice
def multiple():
    n=3
    while True:
        yield n
        n+=3
r=list(islice(multiple(),5))
print(r)
```

```python
def even():
    n=2
    while True:
        yield n
        n+=2
c=0
for num in even():
    print(num,end=" ")
    c+=1
    if c==10:
        break
```
