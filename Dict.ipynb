```python
n=int(input())
d={}
for i in range(n):
    k=input("Enter key:")
    v=int(input("Enter values:"))
    d[k]=v
print(d)
print(d.items())
print(d.values())
print(d.keys())
print(dict(sorted(d.items(),key=lambda x:x[1])))
print(max(d,key=d.get))
print(min(d))
print(len(d))
print(dict(sorted(d.items(),key=lambda x:x[-1])))
print(max(d.items()))
```

```python
d = {"a": 10, "b": 25, "c": 5}

max_key = max(d, key=d.get)
print("Max value key:", max_key)
```

```python
d = {"a": 3, "b": 1, "c": 2}

sorted_d = dict(sorted(d.items(),key=lambda x:x[1],reverse=True))
print(sorted_d)
print(d.pop('a'))
del d['b']
print(d)
print(d.popitem())
print(d)
```
