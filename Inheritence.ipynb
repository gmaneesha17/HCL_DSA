```python
#inheritence-single
class A:
    def show(self):
        print("Parent class")
class B(A):
    def display(self):
        print("Child class")
b=B()
b.show()
b.display()
```

```python
#Multi-level
class A:
    def show(self):
        print("Grand parent")
class B(A):
    def display(self):
        print("Parent")
class C(B):
    def dis(self):
        print("Child")
c=C()
c.show()
c.display()
c.dis()
```

```python
#multiple
class mother:
    def show(self):
        print("Mother")
class father:
    def dis(self):
        print("Father")
class child(mother,father):
    def display(self):
        print("Child")
c=child()
c.show()
c.dis()
c.display()
```

```python
#hierachical:
class A:
    def show(self):
        print("Parent")
class c1(A):
    def dis(self):
        print("Child 1")
class c2(A):
    def display(self):
        print("Child 2")
a=c1()
b=c2()
a.show()
a.dis()
b.show()
b.display()
```

```python
#polymorphism- method overloading-compile-time -same method but diff parameters
class A:
    def add(self,*args):
        print(sum(args))
    # def add(x,y,z):
    #     print(x+y+z)
b=A()
b.add(2,3)
b.add(2,3,4)
```

```python
#method overriding- run time- same method, parameters but in diff classes
class A:
    def show(self):
        print("Parent")
class B(A):
    def show(self):
        print("child")
b=B()
b.show()
```

```python
#overriding-real example- app login system
class user:
    def __init__(self,username):
        self.username=username
        print(self.username)
        print("User created")
class admin(user):
    def __init__(self,username):
        super().__init__(username)
        print(self.username)
        print("Admin")
a=admin("Maneesha")
```

```python
#super-parent class methods, constructors
class A:
    def __init__(self):
        print("Parent")
class B(A):
    def __init__(self):
        super().__init__()
        print("child")
b=B()
```

```python
#Encapsulation-hiding data- banking
class Balance:
    def __init__(self,b):
        self.b=b
    def deposit(self,a):
        self.b+=a
    def show(self):
        print(self.b)
b=Balance(1500)
b.deposit(500)
b.show()
```

```python
#inheritence- vehicle system
class vehicle:
    def show(self):
        print("vehicle starts")
class car(vehicle):
    def dis(self):
        print("car drives")
c=car()
c.show()
c.dis()
```

```python
# method resolution order
class A:
    def show(self):
        print("A")
class B(A):
    def show(self):
        print("B")
class C(B):
    pass
c=C()
c.show()
print(C.mro())
```

```python
#Duck typing philosophy- consider behavoir but not type
class Phonepe:
    def pay(self):
        print("Payed-phonepe")
class googlepay:
    def pay(self):
        print("Payed- Googlepay")
def payment(app):
    app.pay()
payment(Phonepe())
payment(googlepay())
```

```python
# operator overloading
class A:
    def __init__(self,x):
        self.x=x
    def __add__(self,y):
        return self.x+y.x
a=A(10)
b=A(20)
print(a+b)
```
