```python
def printcq():
    s=int(input())
    q=int(input())
    cq=[None]*s
    front=-1
    rear=-1
    for i in range(q):
        op=input().split()
        if op[0]=="enqueue":
            x=int(op[1])
            if (rear+1)%s==front:
                print("Queue full")
            else:
                if front==-1:
                    front=0
                rear=(rear+1)%s
                cq[rear]=x
        elif op[0]=="dequeue":
            if front==-1:
                print("queue empty")
            else:
                cq[front]=None
                if front==rear:
                    front=rear=-1
                front=(front+1)%s
        elif op[0]=="peak":
            if front!=-1:
                print(cq[front])
        elif op[0]=="printqueue":
            print(cq)
printcq()
```
