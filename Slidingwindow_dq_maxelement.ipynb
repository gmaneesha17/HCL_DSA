```python
from collections import deque

def sliding_window_max(A, k):
    dq = deque()
    res = []

    for i in range(len(A)):
        
        while dq and dq[0] <= i - k:
            dq.popleft()

        while dq and A[dq[-1]] < A[i]:
            dq.pop()

        dq.append(i)

        if i >= k - 1:
            res.append(A[dq[0]])

    return res


A = list(map(int, input().split()))
k = int(input())

print(sliding_window_max(A, k))
```
