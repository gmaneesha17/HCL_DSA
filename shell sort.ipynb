```python
def shell(a):
    n=len(a)
    gap=n//2
    while gap>0:
        for i in range(gap,n):
            t=a[i]
            j=i
            while j>=gap and a[j-gap]>t:
                a[j]=a[j-gap]
                j-=gap
            a[j]=t
        gap=gap//2
    return a
a=list(map(int,input().split()))
print(shell(a))
    
```
