```python
def first(a,t):
    l,h=0,len(a)-1
    f=-1
    while l<=h:
        m=(l+h)//2
        if a[m]==t:
            f=m
            h=m-1
        elif a[m]<t:
            l=m+1
        else:
            h=m-1
    return f
def last(a,t):
    l,h=0,len(a)-1
    la=-1
    while l<=h:
        m=(l+h)//2
        if a[m]==t:
            la=m
            l=m+1
        elif a[m]<t:
            l=m+1
        else:
            h=m-1
    return la
a=list(map(int,input().split()))
t=int(input())
f=first(a,t)
l=last(a,t)
print("Result:",[f,l])

```
