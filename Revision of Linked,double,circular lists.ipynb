```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class circularlinkedlist:
    def __init__(self):
        self.head=None
    def insert_end(self,data):
        nn=Node(data)
        if self.head is None:
            self.head=nn
            nn.next=self.head
            return
        t=self.head
        while t.next!=self.head:
            t=t.next
        t.next=nn
        nn.next=self.head
    def insert_beg(self,data):
        nn=Node(data)
        if self.head is None:
            self.head=nn
            nn.next=self.head
            return
        t=self.head
        while t.next!=self.head:
            t=t.next
        nn.next=self.head
        t.next=nn
        self.head=nn
    def insert_pos(self,data,pos):
        nn=Node(data)
        if self.head is None:
            self.head=nn
            nn.next=self.head
            return
        c=1
        t=self.head
        if pos==1:
            self.insert_beg(data)
            return
        while t.next!=self.head and c<pos-1:
            c+=1
            t=t.next
        if not t:
            print("Invalid")
            return
        nn.next=t.next
        t.next=nn
    def delete_end(self):
        if self.head is None:
            print("empty")
            return
        if self.head.next == self.head:
            self.head = None
            return
        t=self.head
        while t.next.next!=self.head:
            t=t.next
        t.next=self.head
    def delete_beg(self):
        if self.head is None:
            print("empty")
            return
        t=self.head
        if self.head.next==self.head:
            self.head=None
            return
        while t.next!=self.head:
            t=t.next
        self.head=self.head.next
        t.next=self.head
        
    def delete_pos(self,value):
        if self.head is None:
            print("empty")
            return
        t=self.head
        if self.head.data==value:
            self.delete_beg()
            return
        prev=None
        while t.next!=self.head and t.data!=value:
            prev=t
            t=t.next
        if t.data!=value:
            print("Invalid")
            return
        prev.next=t.next
    def display(self):
        if self.head is None:
            print("empty")
            return
        t=self.head
        while t:
            print(t.data,end=" ")
            t=t.next
            if t==self.head:
                print("Back to head")
                break
ll=circularlinkedlist()
while True:
    print("\n--- Circular LINKED LIST MENU ---")
    print("1. Insert Beginning")
    print("2. Insert End")
    print("3. Insert Position")
    print("4. Delete Beginning")
    print("5. Delete End")
    print("6. Delete Value")
    print("7. Display")
    print("8. Exit")

    ch = int(input("Enter choice: "))

    if ch == 1:
        val = int(input("Enter value: "))
        ll.insert_beg(val)

    elif ch == 2:
        val = int(input("Enter value: "))
        ll.insert_end(val)

    elif ch == 3:
        val = int(input("Enter value: "))
        pos = int(input("Enter position: "))
        ll.insert_pos(val, pos)

    elif ch == 4:
        ll.delete_beg()

    elif ch == 5:
        ll.delete_end()

    elif ch == 6:
        val = int(input("Enter value: "))
        ll.delete_pos(val)

    elif ch == 7:
        ll.display()

    elif ch == 8:
        print("Exit")
        break

    else:
        print("Invalid choice")
        
        
            
```
