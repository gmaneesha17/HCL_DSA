```python
def anagram(s1,s2):
    if len(s1)!=len(s2):
        return False
    fre={}
    for ch in s1:
        if ch in fre:
            fre[ch]+=1
        else:
            fre[ch]=1
    for ch in s2:
        if ch not in fre:
            return False
        fre[ch]-=1
        if fre[ch]<0:
            return False
    return True
s1=input()
s2=input()
print(anagram(s1,s2))
```
