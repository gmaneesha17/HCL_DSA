```python
def hello():
    print("I am maneesha")
hello()
```

```python
def sum(a,b):
    return a+b
print(sum(1,2))
print(sum(2,3))
```

```python
# calling or invoking fn
def hi():
    print("I am ....")
hi()   #calling
```

```python
# no args, no return
def add():
    print(5+2)
add()
```

```python
#args, no return
def sub(a,b):
    print(a-b)
sub(7,6)
```

```python
# args,return
def multi(a,b):
    return a*b
print(multi(2,3))
```

```python
# no args, return
def get():
    return 10
print(get())
```

```python
# returning multiple values
def cal(a,b):
    return a+b,a-b
print(cal(6,3))
```

```python
def cal(a,b):
    return a+b,a-b
x,y=cal(9,5)
print(x,y)
```

```python
# assign fn to variable
def hello():
    print("HI")
    return 10
f=hello()
print(f)
```

```python
# passed a fn as a variable
def hi():
    print("HELLO")
    return 10
def call(fun):
    # fun()
    print(fun())
call(hi)
```

```python
#returning fn from another fn
def outer():
    def inner():
        print("HI")
        return 9
    return inner
f=outer()
f()
```

```python
def f1():
    print("A")
def f2():
    print("B")
l=[f1,f2]
l[0]()
l[1]()
```

```python
# default args
def student(x,y=10):
    print(x,y)
student(20)
```

```python
#keyword args
def student(n,a):
    print(n,a)
student(a=21,n="Mani")
```

```python
#position args
def student(n,a):
    print(n,a)
student("Mani",21)
```

```python
#variable args-*args:
def total(*nums):
    print(nums)
total(1,2,3)
```

```python
# **kwargs
def info(**d):
    print(d)
info(a=21,n="Mani")
```

```python
#pass keyword
def fun():
    pass
print(fun())
```

```python
#nested fn
def f1():
    def f2():
        print("F2")
    return f2
f=f1()
print(f())
```

```python
   #gloabl
x=10
def change():
    global x
    x=20
    print(x)
change()
print(x)
```

```python
def display(l):
    for i in l:
        print(i)
display([1,2,3,4,5,6,7,8])
```

```python
# lambda
add=lambda x,y:x+y
print(add(2,3))
```

```python
#map
l=[1,2,3]
r=list(map(lambda x:x*2,l))
print(r)
```

```python
#filter
l=[1,2,3,4]
r=list(filter(lambda x:x%2==0,l))
print(r)
```

```python
#reduce
from functools import reduce
l=[1,2,3,4]
r=reduce(lambda x,y:x+y,l)
print(r)
```
