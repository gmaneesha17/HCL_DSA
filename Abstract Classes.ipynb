```python
from abc import ABC, abstractmethod
class PaymentGateway(ABC):
    @abstractmethod
    def pay(self,amount):
        pass
    @abstractmethod
    def refund(self,amount):
        pass
class CreditCardPayment(PaymentGateway):
    def pay(self,amount):
        print("Payed using CreditCard",amount)
    def refund(self,amount):
        print("Refunded using CreditCard",amount)
class UPIPayment(PaymentGateway):
    def pay(self,amount):
        print("Payed using UPI",amount)
    def refund(self,amount):
        print("Refunded using UPI",amount)
c=CreditCardPayment()
c.pay(1000)
c.refund(500)
u=UPIPayment()
u.pay(1500000)
u.refund(10000)

    
```

```python
from abc import ABC, abstractmethod
class Employee(ABC):
    def __init__(self,name,id):
        self.name=name
        self.id=id
    @abstractmethod
    def calculate_salary(self):
        pass
class FullTimeEmployee(Employee):
    def __init__(self,name,id,salary):
        self.salary=salary
        super().__init__(name,id)
    def calculate_salary(self):
        print(self.salary)
class Inten(Employee):
    def __init__(self,name,id,stiphend):
        self.stiphend=stiphend
        super().__init__(name,id)
    def calculate_salary(self):
        print(self.stiphend)
f=FullTimeEmployee("Mani",100,50000)
i=Inten("ABC",50,15000)
print(f.name,f.id,f.salary)
print(i.name,i.id,i.stiphend)
f.calculate_salary()
i.calculate_salary()
```

```python
from abc import ABC, abstractmethod
class FileHandler(ABC):
    @abstractmethod
    def read(self):
        pass
    @abstractmethod
    def write(self,data):
        pass
class TextFileHandler(FileHandler):
    def read(self):
        print("Read")
    def write(self,data):
        print(data)
class CSVFileHandler(FileHandler):
    def read(self):
        print("read")
    def write(self,data):
        print(data)
t=TextFileHandler()
t.read()
t.write("Hello")
c=CSVFileHandler()
c.read()
c.write("Hi")
    
```

```python
from abc import ABC, abstractmethod
import re
class Validator(ABC):
    @abstractmethod
    def validate(self,data):
        pass
class EmailValidator(Validator):
    def validate(self,data):
        p=r"^[\w\.-]+@[\w\.-]+\.\w+$"
        if re.match(p,data):
            print("Valid Email")
        else:
            print("Invalid Email")
class PasswordValidator(Validator):
    def validate(self,data):
        if (len(data)>=8 and  any(c.isdigit() for c in data) and any(c.isupper() for c in data)):
            print("Correct password")
        else:
            print("Invalid Password")
e=EmailValidator()
e.validate("gmaneesha17@gmail.com")
p=PasswordValidator()
p.validate("Manisha@17")

        
```
