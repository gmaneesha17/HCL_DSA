```python
def peak(a):
    l, h = 0, len(a) - 1

    while l < h:
        m = (l + h) // 2

        if a[m] < a[m + 1]:
            l = m + 1
        else:
            h = m   # ✅ fixed

    return l


a = list(map(int, input().split()))

idx = peak(a)
print("Peak Element:", a[idx])
print("Index:", idx)
```
