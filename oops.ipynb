```python
class student:
    def __init__(self,name):
        self.name=name
s=student("Mani")
print(s.name)
```

```python
#instance method
class student:
    def __init__(self,name):
        self.name=name
    def show(self):
        print(self.name)
s=student("Maneesha")
s.show()
```

```python
#class method
class student:
    c="ABC"
    @classmethod
    def show(cls):
        print(cls.c)
student.show()
        
```

```python
#static method
class math:
    def add(a,b):
        return a+b
print(math.add(2,3))
```

```python
#attributes-class
class student:
    c="Abc"
s1=student()
s2=student()
print(s1.c)
print(s2.c)
```

```python
#attribute-instance
class student:
    def __init__(self,n):
        self.n=n
a=student("AB")
b=student("CD")
print(a.n)
print(b.n)
```

```python
#attribute-local
class student:
    def show():
        x=10
        print(x)
student.show()
```

```python
#attribute-global
x=100
class student:
    def show():
        x=10
        print(x)
student.show()
print(x)
```

```python
#passing members of one class to another class members means methods or variables
class A:
    def show(self):
        print("HI")
class B:
    def display(self,obj):
        print(obj.show())
a=A()
b=B()
b.display(a)
```

```python
#inner class
class outer:
    def __init__(self,n):
        self.n=n
        self.lap=self.lap()
    def show(self):
        print(self.n)
        self.lap.display()
    class lap:
        def display(self):
            print("Laptop")
a=outer("Mani")
a.show()
```

```python
class a:
    def __str__(self):
        return "H hh"
print(a())
```

```python
class a:
    def __repr__(self):
        return "object a"
print(a())
```

```python
class b:
    def __del__(self):
        return "delete"
print(b())
```

```python
class a:
    def __init__(self,x):
        self.x=x
    def __add__(self,y):
        return self.x+y.x
A=a(10)
B=a(30)
print(A+B)
```

```python
class A:
    def __init__(self,x):
        self.x=x
    def __eq__(self,y):
        return self.x==y.x
a=A(10)
b=A(10)
print(a==b)
```
