```python
def merge(a):
    if len(a)>1:
        m=len(a)//2
        left=a[:m]
        right=a[m:]
        merge(left)
        merge(right)
        i=j=k=0
        while i<len(left) and j<len(right):
            if left[i]<right[j]:
                a[k]=left[i]
                i+=1
            else:
                a[k]=right[j]
                j+=1
            k+=1
        while i<len(left):
            a[k]=left[i]
            k+=1
            i+=1
        while j<len(right):
            a[k]=right[j]
            k+=1
            j+=1
    return a
a=list(map(int,input().split()))
print(merge(a))
```
