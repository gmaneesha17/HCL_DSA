```python
def low(text):
    return text.strip().lower()
def apply_twice(func,x):
    return func(func(x))
s=input()
print(apply_twice(low,s))
```

```python
def make_multiplier(k):
    def multiplier(x):
        return x*k
    return multiplier
k=int(input())
double=make_multiplier(k)
print(double(3))
```

```python
def counter():
    count=0
    def next():
        nonlocal count
        count+=1
        return count
    return next
c=counter()
print(c())
print(c())
print(c())
print(c())
```

```python
def make_accumulator(start=0):
    t=start
    def add(value):
         nonlocal t
         t+=value
         return t
    return add
value=int(input())
d=make_accumulator(100)
print(d(value))
    
```
