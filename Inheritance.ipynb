```python
class writer:
    def write(self):
        print("writer class")
class reader():
    def read(self):
        print("Reader class")
class Editor(writer,reader):
    def edit(self):
        print("Editor class")
e=Editor()
e.write()
e.read()
e.edit()
```

```python
class employee():
    def __init__(self,salary,bonus):
        self._salary=salary
        self.__bonus=bonus
    def show_bonus(self):
        print(self.__bonus)
class manager(employee):
    def get_salary(self):
        print("Manager salary:",self._salary)
    def get_bonus(self):
        try:
            print(self.__bonus)
        except AttributeError:
            print("Error")
m=manager(1000,500)
m.show_bonus()
m.get_salary()
m.get_bonus()
            
```

```python
class Logger():
    def log(self):
        print("Logger class")
class FileLogger(Logger):
    def log(self):
        super().log()
        print("FileLogger class")
f=FileLogger()
f.log()
        
```

```python
class vehicle():
    def __init__(self,brand):
        self.brand=brand
    def start(self):
        print(self.brand)
        print("vehicle class")
class car(vehicle):
    def __init__(self,brand,fuel):
        super().__init__(brand)
        self.fuel=fuel
    def ftype(self):
        print(self.fuel)
        print("Car Class")
class electriccar(car):
    def __init__(self,brand,capacity):
        super().__init__(brand,fuel="Electric")
        self.capacity=capacity
    def capa(self):
        print(self.capacity)
        print("Electric class")
e=electriccar("Tesla",75)
e.start()
e.ftype()
e.capa()
```
