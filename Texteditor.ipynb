```python
text = ""
undo = []
redo = []

while True:
    print("\n1. Type")
    print("2. Undo")
    print("3. Redo")
    print("4. Show")
    print("5. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        word = input("Enter text: ")
        undo.append(text)
        text = text + word
        redo.clear()

    elif choice == 2:
        if undo:
            redo.append(text)
            text = undo.pop()
        else:
            print("Nothing to undo")

    elif choice == 3:
        if redo:
            undo.append(text)
            text = redo.pop()
        else:
            print("Nothing to redo")

    elif choice == 4:
        print("Current Text:", text)

    elif choice == 5:
        break

    else:
        print("Invalid choice")
```
