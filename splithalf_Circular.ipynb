```python
class Node:
    def __init__(self, data):
        self.data=data
        self.next=None
def create(a):
    h=None
    t=None
    for i in a:
        new=Node(i)
        if h is None:
            h=new
            t=new
        else:
            t.next=new
            t=new
    t.next=h
    return h
def split(h):
    if h is None or h.next==h:
        return h, None
    slow=h
    fast=h
    while fast.next!=h and fast.next.next!=h:
        slow=slow.next
        fast=fast.next.next
    if fast.next.next==h:
        fast=fast.next
    h1=h
    h2=slow.next
    slow.next=h1
    fast.next=h2
    return h1,h2
def printlist(h):
    t=h
    while t:
        print(t.data)
        t=t.next
        if t==h:
            break
    print("back to head")
a=list(map(int,input().split()))
h=create(a)
h1,h2=split(h)
printlist(h1)
printlist(h2)
    
```
