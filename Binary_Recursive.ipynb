```python
def recursive_binary(a,l,r,t):
    if l>r:
        return -1
    m=(l+r)//2
    if a[m]==t:
        return m
    elif a[m]<t:
        return recursive_binary(a,m+1,r,t)
    else:
        return recursive_binary(a,l,m-1,t)
a=list(map(int,input().split()))
t=int(input())
r=recursive_binary(a,0,len(a)-1,t)
if r!=-1:
    print("Element found at index ",r)
else:
    print("No element found")
```
