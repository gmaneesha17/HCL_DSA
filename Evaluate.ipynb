```python
def evaluate(t):
  st=[]
  for ch in t:
    if ch not in "+-*%":
      st.append(int(ch))
    else:
      b=st.pop()
      a=st.pop()
      if ch=="+":
        st.append(a+b)
      elif ch=="-":
        st.append(a-b)
      elif ch=="*":
        st.append(a*b)
      elif ch=="%":
        st.append(a%b)
  return st[0]
t=input("Enter postfix expression (space-separated tokens): ").split()
print(evaluate(t))
```
