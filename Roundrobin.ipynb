```python
from collections import deque

n = int(input("Enter number of processes: "))
processes = list(map(int, input("Enter burst times: ").split()))
tq = int(input("Enter time quantum: "))

q = deque()
for i in range(n):
    q.append((i+1, processes[i]))

time = 0

while q:
    pid, bt = q.popleft()

    if bt > tq:
        time += tq
        bt -= tq
        print(f"P{pid} runs for {tq} units, remaining = {bt}")
        q.append((pid, bt))
    else:
        time += bt
        print(f"P{pid} completes at time {time}")
```
