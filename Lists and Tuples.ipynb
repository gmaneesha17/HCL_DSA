```python
def listtotuple(s):
    r=[]
    for i in range(0,len(s)-1,2):
        if i+1<len(s):
            r.append((s[i],s[i+1]))
    return r
s=input().split()
print(listtotuple(s))
```

```python
def lastoccu(a):
    r=[]
    for i in range(len(a)-1,0,-1):
        if a[i] not in r:
            r.append(a[i])
    return r[::-1]
a=list(map(int,input().split()))
print(lastoccu(a))
```

```python
def seconlarge(a):
    if len(a)<2:
        return -1
    f=float('-inf')
    s=float('-inf')
    for n in a:
        if n>f:
            s=f
            f=n
        elif n>s and n!=f:
            s=n
    if s==float('-inf'):
        return -1
    for i in range(len(a)):
        if a[i]==s:
            return i
a=list(map(int,input().split()))
print("Index ",seconlarge(a))
```

```python
def reverse(a,s,e):
    while s<e:
        a[s],a[e]=a[e],a[s]
        s+=1
        e-=1
def rotate(a,k):
    n=len(a)
    if k>n:
        k=k%n
    reverse(a,0,n-1)
    reverse(a,0,k-1)
    reverse(a,k,n-1)
    return a
a=list(map(int,input().split()))
k=int(input())
print("Right rotation ",rotate(a,k))
```

```python
def interleave(a,b):
    n=len(a)
    m=len(b)
    r=[]
    l=n if n<m else m
    for i in range(l):
        r.append(a[i])
        r.append(b[i])
    return r
a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(interleave(a,b))
```

```python
def sqeven(a):
    r=[]
    for n in a:
        if n%2==0:
            r.append(n*n)
    return r
a=list(map(int,input().split()))
print(sqeven(a))
```

```python
def cumsum(a):
    t=0
    r=[]
    for i in a:
        if i<0:
            break
        t=t+i
        r.append(t)
    return r
a=list(map(int,input().split()))
print(cumsum(a))
```

```python
def rack(a):
    r=[]
    cs=0
    for i in a:
        if i<=0:
            cs=0
        else:
            if i>cs:
                cs=i
        r.append(cs)
    return r
a=list(map(int,input().split()))
print(rack(a))
    
```
