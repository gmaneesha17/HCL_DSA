```python
def ropes(a,k):
    l=1
    h=max(a)
    ans=0
    while l<=h:
        m=(l+h)//2
        p=0
        for i in a:
            p+=i//m
        if p>=k:
            ans=m
            l=m+1
        else:
            h=m-1
    return ans
a=list(map(int,input().split()))
k=int(input())
print(ropes(a,k))
        
    
```
