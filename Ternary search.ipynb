```python
def ternary(a,t):
    l=0
    h=len(a)-1
    while l<=h:
        m1=l+(h-l)//3
        m2=h-(h-l)//3
        if a[m1]==t:
            return m1
        if a[m2]==t:
            return m2
        if t<a[m1]:
            h=m1-1
        elif t>a[m2]:
            l=m2+1
        else:
            l=m1+1
            h=m2-1
    return -1
a=list(map(int,input().split()))
t=int(input())
print(ternary(a,t))
```
