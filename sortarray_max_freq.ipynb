```python
def sortfreq(a):
    freq={}
    for ch in a:
        if ch in freq:
            freq[ch]+=1
        else:
            freq[ch]=1
    while freq:
        mele=None
        mfreq=-1
        for k in freq:
            if freq[k]>mfreq:
                mfreq=freq[k]
                mele=k
        for _ in range(mfreq):
            print(mele,end=" ")
        del freq[mele]
a=list(map(int,input().split()))
print(sortfreq(a))
```
