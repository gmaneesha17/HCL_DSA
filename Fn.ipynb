```python
def normalize_name(s: str) -> str:
    h_c=False
    for ch in s:
        if ch!=" ":
            h_c=True
            break
    if not h_c:
        return ""
    i=0
    n=len(s)
    r=""
    while i<n and s[i]==" ":
        i+=1
    h_w=True
    while i<n:
        ch=s[i]
        if ch==" ":
            while i<n and s[i]==" ":
                i+=1
            if i<n:
                r+=" "
                h_w=True
        else:
            if h_w:
                if 'a'<=ch<='z':
                    ch=chr(ord(ch)-32)
                r+=ch
                h_w=False
            else:
                if 'A'<=ch<='Z':
                    ch=chr(ord(ch)+32)
                r+=ch
            i+=1
    return r
s=input()
print(normalize_name(s))
                
```

```python
def aggregate(*nums, op="sum"):
    ma=nums[0]
    t=0
    n=len(nums)
    if len(nums)==0:
        return None
    for i in nums:
        t+=i
        if i>ma:
            ma=i
    if op=="sum":
        return t
    elif op=="avg":
        return t/n
    elif op=="max":
        return ma
    else:
        return "Invalid"
n=input().split()
r=[]
for i in n:
    r.append(int(i))
operation=input()
res=aggregate(*r,op=operation)
print(res)
        
        
        
    
```

```python
def make_counter(start=0):
    count=start
    def counter():
        nonlocal count
        count+=1
        return count
    return counter
st=int(input())
c=make_counter(st)
print(c())
print(c())
```

```python
def check_strength(pwd: str, min_len=8) -> dict:
    hu=False
    hl=False
    hd=False
    hs=False
    for ch in pwd:
        if 'A'<=ch<'Z':
            hu=True
        elif 'a'<=ch<='z':
            hl=True
        elif '0'<=ch<'9':
            hd=True
        else:
            hs=True
    length_ok=len(pwd)>=min_len
    s=0
    if hu:
        s+=1
    if hl:
        s+=1
    if hs:
        s+=1
    if length_ok:
        s+=1
    if hd:
        s+=1
    return {"length_ok":length_ok,"h_u":hu,"h_l":hl,"h_d":hd,"h_s":hs,"score":s}
pwd=input()
r=check_strength(pwd)
print(r)
```
