```python
import unicodedata
from collections import defaultdict
def normalize(text):
    text=text.casefold()
    text=unicodedata.normalize('NFKD',text)
    text=''.join(c for c in text if not unicodedata.combining(c))
    return text
def anagram(words):
    g=defaultdict(list)
    for i in words:
        norm=normalize(i)
        key=''.join(sorted(norm))
        g[key].append(i)
    return list(g.values())
n=int(input())
w=[]
for _ in range(n):
    w.append(input())
r=anagram(w)
for i in r:
    print(i)
```
