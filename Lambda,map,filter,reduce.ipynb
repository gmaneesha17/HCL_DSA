```python
from functools import reduce
n=[900,200,30]
d=list(map(lambda x:x-(x*0.10),n))
f=list(filter(lambda x:x>=200,d))
r=reduce(lambda a,b:a+b,f,0)
print(d)
print(f)
print(r)
```

```python
from functools import reduce
s=[1,2,3,4]
w=[10,20,30]
l=list(map(lambda x,y:x*y,s,w))
r=reduce(lambda a,b:a+b,l,0)
print(l)
print(r)
```
