```python
def rotate(a,k):
    l=0
    h=len(a)-1
    while l<=h:
        m=(l+h)//2
        if a[m]==k:
            return m
        if a[l]<=a[m]:
            if a[l]<=k<a[m]:
                h=m-1
            else:
                l=m+1
        else:
            if a[m]<k<=a[h]:
                l=m+1
            else:
                h=m-1
    return -1
a=list(map(int,input().split()))
k=int(input())
r=rotate(a,k)
if r!=-1:
    print("Element found at :", r)
else:
    print("Not Found")
```
